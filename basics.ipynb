```python
%matplotlib inline 
#magic function starts with %
import matplotlib.pyplot as plt
plt.plot(5,5,'o')
plt.show()
```

```python
x=[10,20,30,40]
y=[20,25,35,55]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.title('Linear graph')#set title
plt.plot(x, y)
plt.show()
```

```python
plt.title("Linear graph",fontsize="25" ,color="red")
plt.plot(x,y)
plt.show()
```

```python
plt.title("Linear graph",fontsize="25" ,color="red")
plt.plot(x,y)
plt.xlabel('x=axis',color="green")
plt.ylabel('y-axis',color="Blue")
plt.show()
```

```python
plt.title("Linear graph",fontsize="25" ,color="red")
plt.plot(x,y,color="red")
plt.xlabel('x=axis',color="green")
plt.ylabel('y-axis',color="Blue")
plt.show()
```
